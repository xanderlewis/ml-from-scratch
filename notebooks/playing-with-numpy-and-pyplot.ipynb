# Experimenting a little bit with Jupyter notebooks.

...and playing with generating stuff in numpy and plotting with matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Let's play with covariance matrices.

```python
points = np.random.multivariate_normal(mean=[0,0], cov=[[1,1],[1,1]], size=100)
```

```python
points.shape
```

```python
plt.scatter(points[:, 0], points[:, 1], s=2, c='red')
```

Perfectly correlated variables!

```python
points = np.random.multivariate_normal(mean=[0,0], cov=[[1, 0.8],[0.8, 1]], size=1000)
plt.scatter(points[:, 0], points[:, 1], c='hotpink', s=3)
```

Slightly less correlated variables. Each is 100% correlated with itself, but only 80% (?) with the other.

```python
points = np.random.multivariate_normal(mean=[0,0], cov=[[1, 0],[0, 1]], size=1000)
plt.scatter(points[:, 0], points[:, 1], c='hotpink', s=3)
```

Entirely independent variables.

Can I plot a 3D distribution like this?

```python
# Generate a point cloud
points = np.random.multivariate_normal(mean=[0,0,0], cov=np.identity(3), size=1000)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=2, c='hotpink')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
points.shape
```

```python
# Generate (another) point cloud
points = np.random.multivariate_normal(mean=[0,0,0], cov=[[1, 0.8, 0.8],[0.8, 1, 0.8],[0.8, 0.8, 1]], size=1000)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=2, c='hotpink')
```

Note: our matrix here is determined by only three of its entries: the correlations of x with y, x with z, and y with z.

```python
points = np.random.multivariate_normal(mean=[0,0,0], cov=[[1, -0.95, 0],[-0.95, 1, 0],[0, 0, 1]], size=1000)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=1, c='darkorange')
```
